```python
def process_irs_outflow(df):
    df = df[df.State_Code_Origin == 6]
    df = df[df.State_Code_Dest == 6]


    #filter out non migrants unneded data
    df=df[df["County_Name"].str.contains("Tot Mig|Non-Mig|Total Mig|Other Flows|Foreign")==False]

    #filter out non bay area counties
    df=df.loc[df.County_Code_Origin.isin(code_list)]
    df=df.loc[df.County_Code_Dest.isin(code_list)]
    
    df = df[df['County_Code_Origin'] != df['County_Code_Dest']]


    #aggregate outflows using groupby
    outflows=df.groupby(['County_Code_Origin']).sum()
    outflows=outflows.drop(columns=['County_Code_Dest','State_Code_Origin','State_Code_Dest'])

    outflows['Name']=name_list
    return outflows
```

```python
def process_irs_outflow_xls(df1):

    df1.drop([0,1,2,3,4,5], inplace = True )
    df1.drop(df1.columns[[4]], axis = 1, inplace = True)

    df1.rename(columns={ df1.columns[0]: "State_Code_Origin" }, inplace = True)
    df1.rename(columns={ df1.columns[1]: "County_Code_Origin" }, inplace = True)
    df1.rename(columns={ df1.columns[2]: "State_Code_Dest" }, inplace = True)
    df1.rename(columns={ df1.columns[3]: "County_Code_Dest" }, inplace = True)
    df1.rename(columns={ df1.columns[4]: "County_Name" }, inplace = True)
    df1.rename(columns={ df1.columns[5]: "Return_Num" }, inplace = True)
    df1.rename(columns={ df1.columns[6]: "Exmpt_Num" }, inplace = True)
    df1.rename(columns={ df1.columns[7]: "Aggr_AGI" }, inplace = True)

    df1 = df1[df1.State_Code_Origin == '06']
    df1 = df1[df1.State_Code_Dest =='06']

    code_list_str = ['001', '013', '041', '047', '055', '069', '075', '077', '081', '085', '087', '095', '097', '099']

    
    df1=df1[df1["County_Name"].str.contains("Tot Mig|Non-Mig|Total Mig|Other Flows|Foreign|Region|Non Mig|Same State|Surpress")==False]



    df1 = df1[df1['County_Code_Origin'].isin(code_list_str)==True]
    df1 = df1[df1['County_Code_Dest'].isin(code_list_str)==True]


    df1 = df1[df1['County_Code_Origin'] != df1['County_Code_Dest']]
    df1.drop(df1.columns[[0,2,3,4]], axis = 1, inplace = True)
    
    df1["Return_Num"] = df1["Return_Num"].astype(str).astype(int)
    df1["Exmpt_Num"] = df1["Exmpt_Num"].astype(str).astype(int)
    df1["Aggr_AGI"] = df1["Aggr_AGI"].astype(str).astype(int)



    outflow=df1.groupby(['County_Code_Origin']).sum()
    return outflow
```
